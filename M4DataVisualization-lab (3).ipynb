```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# your code goes here
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

df['ConvertedComp'].plot(kind='hist')
plt.title('Histogram of ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM master
GROUP BY age
ORDER BY age
"""

# Read the entire table into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Plot a boxplot for 'Age' column
plt.figure(figsize=(6, 8))
df['Age'].plot.box()
plt.title('Boxplot of Age Distribution')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT *
FROM master
"""

QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Read 'Age' and 'WorkWeekHrs' columns into a DataFrame using Pandas
df = pd.read_sql_query(QUERY, conn)

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='darkblue')
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('Age')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

# Read 'Age', 'WorkWeekHrs', and 'CodeRevHrs' columns into a DataFrame using Pandas
df = pd.read_sql_query(QUERY, conn)

# Define bubble sizes based on 'Age' column
bubble_sizes = df['Age'] * 5  # Adjust the multiplication factor for appropriate bubble sizes

# Create a bubble plot
plt.figure(figsize=(10, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_sizes, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT* 
FROM DatabaseDesireNextYear
"""
df = pd.read_sql_query(QUERY,conn)

# Assuming you have a DataFrame 'df' with the relevant data
# Extract the 'DatabaseDesireNextYear' column
databases_desired = df['DatabaseDesireNextYear']
print(databases_desired)

# Count the occurrences of each database
database_counts = databases_desired.value_counts().head(5)

# Create a pie chart for the top 5 databases
plt.figure(figsize=(8, 8))
plt.pie(database_counts, labels=database_counts.index, autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, 
       AVG(WorkWeekHrs) AS MedianWorkWeekHrs, 
       AVG(CodeRevHrs) AS MedianCodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
GROUP BY Age
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.bar(df['Age'], df['MedianWorkWeekHrs'], label='Median WorkWeekHrs')
plt.bar(df['Age'], df['MedianCodeRevHrs'], bottom=df['MedianWorkWeekHrs'], label='Median CodeRevHrs')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.legend()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, 
       AVG(ConvertedComp) AS MedianConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.plot(df['Age'], df['MedianConvertedComp'], label='Median ConvertedComp')
plt.xlabel('Age Group')
plt.ylabel('Median Salary')
plt.title('Median Salary for Age Group 45-60')
plt.legend()
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) AS BranchCount
FROM master
GROUP BY MainBranch
"""
df = pd.read_sql_query(QUERY,conn)

plt.figure(figsize=(10, 6))
plt.barh(df['MainBranch'], df['BranchCount'], color='Green')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branches')
plt.show()
```
